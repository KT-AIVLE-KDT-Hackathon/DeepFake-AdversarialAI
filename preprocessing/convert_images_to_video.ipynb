```python
import cv2
import os

# 이미지 폴더 경로
image_folder = 'C:\\Users\\User\\Desktop\\ha\\Input_video_T'
# 출력 비디오 파일 경로
output_video = 'C:\\Users\\User\\Desktop\\ha\\Input_video_T.mp4'

# 이미지 파일 목록 가져오기
images = [img for img in os.listdir(image_folder) if img.endswith(".jpg") or img.endswith(".png")]
images.sort()  # 정렬 (필요에 따라)

# 첫 번째 이미지로 비디오 캡처 객체 초기화
first_image = cv2.imread(os.path.join(image_folder, images[0]))
height, width, layers = first_image.shape

# 비디오 작성을 위한 VideoWriter 객체 생성
fourcc = cv2.VideoWriter_fourcc(*'mp4v')  # 또는 'XVID'
video = cv2.VideoWriter(output_video, fourcc, 30, (width, height))  # 초당 30프레임

# 이미지 파일들을 비디오에 추가
for image in images:
    video.write(cv2.imread(os.path.join(image_folder, image)))

# 비디오 작성을 종료하고 자원 해제
video.release()
cv2.destroyAllWindows()

print("비디오 생성 완료:", output_video)
```
